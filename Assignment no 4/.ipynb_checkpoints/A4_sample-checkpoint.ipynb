```python
import  matplotlib.pyplot as plt
from matplotlib.figure import Figure
```

```python
# initialize the data

x=[10, 20, 30, 40]
y=[20, 25, 35, 55]
```

```python
# create a new figure with width = 5 inches and height = 4 inches
fig = plt.figure(figsize =(5,4))

# create a  first axis for figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# create a  second axis for figure
ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])
```

```python
# plotting the data
plt.plot(x,y)

# add title to the plot
plt.title("Linear graph", fontsize =25, color="green")

# add label to the x-axis and y-axis

plt.xlabel("X Data")
plt.ylabel("Y Data")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# reading the tip.csv

df = pd.read_csv('tips.csv')
df.head()
```

```python
# initialize the data

x=df['day']
y=df['total_bill']

#plotting the data

plt.bar(x,y)
plt.title('Tips Data set')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show
```

```python
# Histogram

plt.hist(x, bins=4, color='green', edgecolor='blue')
plt.ylabel('Frequency')
plt.xlabel('Total Bill')
```

```python
# pie chart

import matplotlib.pyplot as plt

cars = ['Awdi','BMW','FORD','Tesla','Jagwar']
data = [23, 10,35,15, 12]

plt.pie(data, labels=cars)
plt.show()
```

# Catgorical scatterplot
The default representation of the data in catplot() uses a scatterplot.
There are two different scatter plots in seaborn.



```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='box')
```
